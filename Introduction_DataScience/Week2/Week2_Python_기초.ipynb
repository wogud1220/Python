### 1.2 사칙연산

```python
x = 10           # 정수 대입
x1 = x + 4       # 덧셈
x2 = x - 4       # 뺄셈
x3 = x * 4       # 곱셈
x4 = x / 4       # 나눗셈
x5 = x % 4       # 나눗셈(나머지 구하기)

print(x, x1, x2, x3, x4, x5)
print(type(x), type(x1), type(x2), type(x3), type(x4), type(x5))
```

```python
x1 += 4          # x1 = x1+4와 동일
x2 -= 4          # x2 = x2-4와 동일
x3 *= 4          # x3 = x3*4와 동일
x4 /= 4          # x4 = x4/4와 동일
x5 %= 4          # x5 = x5%4와 동일

print(x1, x2, x3, x4, x5)
print(type(x1), type(x2), type(x3), type(x4), type(x5))
```

```python
y = 3.7          # 실수 대입
y1 = y + 4       # 덧셈
y2 = y - 4       # 뺄셈
y3 = y * 4       # 곱셈
y4 = y / 4       # 나눗셈
y5 = y % 4       # 나눗셈(나머지 구하기)

print(y1, y2, y3, y4, y5)
print(type(y1), type(y2), type(y3), type(y4), type(y5))
```

```python
z = 10.0
z1 = z // 2      # integer division
z2 = z ** 2      # square of z
z3 = z ** 0.5   # square root
z4 = pow(z,2)    # equivalent to square of z
z5 = round(z)    # rounding z to its nearest integer
z6 = int(z)      # type casting float to int

print(z, z1, z2, z3, z4, z5, z6)
print(type(z), type(z1), type(z2), type(z3), type(z4), type(z5), type(z6))
```

```python
# 소수점 자리수 정하기
print ('%.2f' % z3)
```

### 1.3 Math 라이브러리 사용

```python
import math

x = 4
print(math.sqrt(x))      # sqrt(4) = 2
print(math.pow(x,2))     # 4**2 = 16
print(math.exp(x))       # exp(4) = 54.6
print(math.log(x,2))     # log based 2  (default is natural logarithm)
print(math.fabs(-4))     # absolute value
print(math.factorial(x)) # 4! = 4 x 3 x 2 x 1 = 24
```

### 1.4 Booleans

```python
t = True
f = False

#### Booleans : True or False를 반환함
#### AND 연산 : 둘 다 True일 때는 True일 때를 제외하고 모두 False를 반환
#### OR 연산 : 둘 중 하나라도 True이면 무조건 True를 반환
#### NOT 연산 : True면 False를, False면 True를 반환
#### XOR 연산 : 같지 않을 경우 True를 반환

print(type(t), type(f))
print(t and f)
print(t or f)
print(not t)
print(t != f)
```

### 1.6 문자열 관련 기타 기능

```python
word = "soft"

#앞글자만 대문자로 출력
print(word.capitalize())

#문자 전체를 대문자로 출력
print(word.upper())

#왼쪽 정렬
#변수: 문자열의 길이, 공백 대신 들어가는 문자
#7개의 문자열 중 왼쪽으로 정렬 후 빈 자리는 #으로 채운다
print(word.rjust(7, "#"))

#중앙 정렬
#변수: 문자열의 길이, 공백 대신 들어가는 문자
print(word.center(20, "#"))

#문자열 치환
print(word.replace('t', 'j'))

#문자열 주변의 공백 문자 삭제
print('  soft '.strip())
```

### 1.9 조건문, 반복문

```python
#조건문 if

student_a = 98
student_b = 50

# C언와 다른 점은 else if가 아닌 elif 로 사용
if student_a > student_b:
    print ("student_a WIN")
elif student_a == student_b:
    print ("draw")
else:
    print ("student_b WIN")
```

```python
# 구구단
for i in range(2,10):
    for j in range(1,10):
        print (i*j, end="\t")
    print (" ")
```

```python
#enumerate는 리스트가 있는 경우 순서와 리스트의 값을 전달하는 기능을 가짐
Fruit = ["Mango", "Apple", "Banana", "Blackberry", "Cherry", "Avocado"]
for idx, fruit in enumerate(Fruit):
  print(idx, fruit)
```

### 1.10 함수

```python
#함수 선언
def sign(x):
    if x > 0:
        return 'positive'
    elif x < 0:
        return 'negative'
    else:
        return 'zero'

#반복문을 통한 함수 호출
for x in [-1, 0, 1]:
    print(sign(x))
```

### 1.11 파일 읽기, 쓰기

```python
#파일 생성
#파일 옵션
# w: 생성, a: 추가, r: 읽기
names = ["이승훈", "최아라", "Usama Ejaz", "Ali Hamza","박태정","박현수"]
fruits = ["Mango", "Apple", "Banana", "Blackberry","Cherry","Avocado"]
reult_str = list()

# open은 파일 입출력을 위한 파일 스트림 형성을 의미함. 여기서는 open으로 스트림을 열고 이를 h라는 변수로 지칭한다는 의미
# csv는 comma seperated value라는 의미로 주로 데이터 사이언스에서 많이 사용되는 형식
with open("data.csv", "w") as h:
    for num in range(0, len(names)):
        save_str = names[num] + ", " + fruits[num]
        print (save_str)
        h.write(save_str + "\n")
```
