```python
import pandas as pd

stu_dict = {'국어':[90, 20, 50],
            '수학':[45, 50, 30],
            '영어':[50, 30, 20]}
stu_data = pd.DataFrame(stu_dict)
print(stu_data)
```

```python
stu_data.index = ['승훈', '알리', '우사마']
print((stu_data))
```

좌측 인덱스행을 변경

```python
stu_data2 = pd.DataFrame(stu_data, columns = ['국어', '수학'])
print(stu_data2)
```

열을 국어, 수학 영영만 나오게함

```python
print(stu_data)
```

```python
stu_data.loc['철수'] = [59, 80, 20]
print(stu_data)
```

넣는 거는 loc, 3개 맞춰서 넣기
삭제는 drop


```python
stu_data = stu_data.drop(['알리'])
print(stu_data)
```

```python
stu_data['과학']= (50, 60, 20)
print(stu_data)
```

```python
stu_data = stu_data.drop(['수학'], axis=1)
print(stu_data)
```

축을 정해주기.
기본으로 0이 되어있음 행을 기준으로 바라보고 아래를 바라보고
1로 하면 열을 바라보고 옆을 바라보고

```python
stu_data.loc['종국'] = [52, 81, 87]
stu_data.loc['재석'] = [54, 85, 48]
stu_data.loc['호동'] = [90, 88, 55]

stu_data.loc['재형'] = [60, 63, 33]
stu_data.loc['윈터'] = [24, 67, 45]
stu_data.loc['리나'] = [55, 77, 39]

print(stu_data)
```

arr[:2, 1:]
x축은 2번째 전까지 y축은 1번째부터 끝까지
3x3에선 우상단 4칸이 출력됨

```python
print(stu_data)
print()
print(stu_data.iloc[:2])
print()
print(stu_data.iloc[:2, :2])
```

```python
print(stu_data.iloc[2:5, 1:])
```
