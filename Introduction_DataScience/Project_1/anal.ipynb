```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded files
temp_file_path = 'temp.csv'
pm_file_path = 'converted_.csv'

temperature_data = pd.read_csv(temp_file_path)
pm_data = pd.read_csv(pm_file_path)

# Convert the 'Day' column to datetime format
temperature_data['Day'] = pd.to_datetime(temperature_data['Day'], format='%Y.%m.%d')

# Extract the relevant data for Songpa from the temperature dataset
songpa_temp = temperature_data[['Day', 'Songpa']]
songpa_temp_summer = songpa_temp[(songpa_temp['Day'].dt.month >= 6) & (songpa_temp['Day'].dt.month <= 8)]

# Calculate the average temperature for June, July, and August
songpa_temp_summer['Month'] = songpa_temp_summer['Day'].dt.month
monthly_avg_temp = songpa_temp_summer.groupby('Month')['Songpa'].mean()

# Extract the relevant data for Songpa from the PM dataset
pm_data.set_index(['City', 'District_1', 'District_2'], inplace=True)
songpa_pm = pm_data.loc[('Seoul', 'Songpa', 'Songpa')][['2023.06', '2023.07', '2023.08']].astype(int)

# Prepare data for plotting
months = ['June', 'July', 'August']
temp_values = [monthly_avg_temp[6], monthly_avg_temp[7], monthly_avg_temp[8]]
pm_values = [songpa_pm['2023.06'], songpa_pm['2023.07'], songpa_pm['2023.08']]

# Plot temperature data as a line graph
plt.figure(figsize=(12, 6))
plt.plot(months, temp_values, marker='o', label='Temperature (C)', color='r')
plt.title('Average Temperature in Songpa (June to August 2023)')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot PM data as a line graph
plt.figure(figsize=(12, 6))
plt.plot(months, pm_values, marker='o', label='PM10 Levels (g/m)',color = 'brown')
plt.title('Average PM10 Levels in Songpa (June to August 2023)')
plt.xlabel('Month')
plt.ylabel('PM10 (g/m)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded files
temp_file_path = 'temp.csv'
pm_file_path = 'converted_.csv'

temperature_data = pd.read_csv(temp_file_path)
pm_data = pd.read_csv(pm_file_path)

# Convert the 'Day' column to datetime format
temperature_data['Day'] = pd.to_datetime(temperature_data['Day'], format='%Y.%m.%d')

# Extract the relevant data for Songpa from the temperature dataset
songpa_temp = temperature_data[['Day', 'Songpa']]
songpa_temp_summer = songpa_temp[(songpa_temp['Day'].dt.month >= 6) & (songpa_temp['Day'].dt.month <= 8)]

# Calculate the average temperature for June, July, and August
songpa_temp_summer['Month'] = songpa_temp_summer['Day'].dt.month
monthly_avg_temp = songpa_temp_summer.groupby('Month')['Songpa'].mean()

# Extract the relevant data for Songpa from the PM dataset
pm_data.set_index(['City', 'District_1', 'District_2'], inplace=True)
songpa_pm = pm_data.loc[('Seoul', 'Songpa', 'Songpa')][['2023.06', '2023.07', '2023.08']].astype(int)

# Prepare data for plotting
months = ['June', 'July', 'August']
temp_values = [monthly_avg_temp[6], monthly_avg_temp[7], monthly_avg_temp[8]]
pm_values = [songpa_pm['2023.06'], songpa_pm['2023.07'], songpa_pm['2023.08']]

# Plot both temperature and PM data on the same figure with dual y-axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot temperature data on the primary y-axis
color = 'red'
ax1.set_xlabel('Month')
ax1.set_ylabel('Temperature (C)', color=color)
ax1.plot(months, temp_values, marker='o', color=color, label='Temperature (C)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

# Create a secondary y-axis for PM data
ax2 = ax1.twinx()
color = 'blue'
ax2.set_ylabel('PM10 Levels (g/m)', color=color)
ax2.plot(months, pm_values, marker='o', color=color, label='PM10 Levels (g/m)')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

# Adding titles and grid
plt.title('Average Temperature and PM10 Levels in Songpa (June to August 2023)')
fig.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the uploaded files
temp_file_path = 'temp.csv'
pm_file_path = 'converted_.csv'

temperature_data = pd.read_csv(temp_file_path)
pm_data = pd.read_csv(pm_file_path)

# Convert the 'Day' column to datetime format
temperature_data['Day'] = pd.to_datetime(temperature_data['Day'], format='%Y.%m.%d')

# Extract the relevant data for Songpa from the temperature dataset
songpa_temp = temperature_data[['Day', 'Songpa']]
songpa_temp_summer = songpa_temp[(songpa_temp['Day'].dt.month >= 6) & (songpa_temp['Day'].dt.month <= 8)]

# Calculate the average temperature for June, July, and August
songpa_temp_summer['Month'] = songpa_temp_summer['Day'].dt.month
monthly_avg_temp = songpa_temp_summer.groupby('Month')['Songpa'].mean()

# Extract the relevant data for Songpa from the PM dataset
pm_data.set_index(['City', 'District_1', 'District_2'], inplace=True)
songpa_pm = pm_data.loc[('Seoul', 'Songpa', 'Songpa')][['2023.06', '2023.07', '2023.08']].astype(int)

# Prepare data for plotting
months = ['June', 'July', 'August']
temp_values = [monthly_avg_temp[6], monthly_avg_temp[7], monthly_avg_temp[8]]
pm_values = [songpa_pm['2023.06'], songpa_pm['2023.07'], songpa_pm['2023.08']]

# Create a dataframe for correlation analysis
correlation_data = pd.DataFrame({
    'Temperature': temp_values,
    'PM10': pm_values
})

# Calculate the correlation coefficient
correlation_coefficient = correlation_data.corr().iloc[0, 1]

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data.corr(), annot=True, cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation between Temperature and PM10 Levels')
plt.show()

print("Correlation Coefficient:", correlation_coefficient)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#   
temperature_data = pd.read_csv("./temp.csv")

# 'Day' datetime    
temperature_data['Day'] = pd.to_datetime(temperature_data['Day'], format='%Y.%m.%d')
temperature_data['Month'] = temperature_data['Day'].dt.month

#  6, 7, 8      
monthly_avg_temp = temperature_data.groupby('Month')['Songpa'].mean().loc[[6, 7, 8]].value

#   
heat_related_patient_counts = [1, 9, 6]  # 6, 7, 8

#   
pm10_data = pd.read_csv("./converted_.csv")

#  6, 7, 8   
songpa_pm10_data = pm10_data[(pm10_data['City'] == 'Seoul') & (pm10_data['District_1'] == 'Songpa')]
songpa_pm10_values = songpa_pm10_data[['2023.06', '2023.07', '2023.08']].values.flatten().astype(float)

#  
data = {
    'Month': [6, 7, 8],
    'Heat-related Patient Count': heat_related_patient_counts,
    'Average Temperature (C)': monthly_avg_temp,
    'Average PM10 (g/m)': songpa_pm10_values
}
df = pd.DataFrame(data)

#  
correlation_matrix = df[['Heat-related Patient Count', 'Average Temperature (C)', 'Average PM10 (g/m)']].corr()

#  
plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation between Heat-related Patient Count, Temperature, and Fine Dust (PM10)')
plt.show()

#  
print("Correlation Coefficient:")
print(correlation_matrix)
```
