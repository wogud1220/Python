```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 샘플 데이터 생성
data = {
    '인구': [10000, 20000, 15000, 30000, 25000, 12000, 22000],
    '밀도': [3000, 4000, 3500, 5000, 4500, 3200, 4200],
    '면적': [10, 20, 15, 30, 25, 12, 22]
}
df = pd.DataFrame(data)

# K-means 클러스터링 수행
kmeans = KMeans(n_clusters=3)
df['클러스터'] = kmeans.fit_predict(df)

# 결과 출력
print(df)

# 3D 그래프로 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['인구'], df['밀도'], df['면적'], c=df['클러스터'], cmap='viridis', marker='o')
ax.set_xlabel('인구')
ax.set_ylabel('밀도')
ax.set_zlabel('면적')
plt.show()
```

```python
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

address = geocoding_reverse('36.5760732781656, 128.15935928504484')
print(address)
```
