### 1.1 데이터 프레임

```python
import pandas as pd

stu_dict = {'국어':[90, 20, 50],
           '수학':[45, 50, 30],
            '영어':[50,30,20]}
stu_data = pd.DataFrame(stu_dict)
print (stu_data)
```

```python
stu_data.index = ['승훈', '알리', '우사마']
print (stu_data)
```

```python
stu_data2 = pd.DataFrame(stu_dict, columns=['국어','수학'])
print (stu_data2)
```

### 1.5 데이터 프레임 합치기, 나누기

```python
stu_data
```

```python
new_student=[{'국어':30, '영어':20, '과학':80}, {'국어':53, '영어':94, '과학':92}]
stu_data2 = pd.DataFrame(new_student, index=['명수', '준호'], columns=['국어','영어','과학'])
print(stu_data2)
```

```python
concat_stu_data = pd.concat([stu_data, stu_data2])
print (concat_stu_data)
```

## 수학 점수 Dataframe을 하나 추가해서 concat_stu_data 와 합치세요

```python
math_score = pd.DataFrame([30, 20, 40, 18, 30, 60, 30, 80], index=['승훈','우사마','철수','종국','재석','호동','명수','준호'] ,columns=["수학"])
print (math_score)
```

```python
concat_stu_data2 = pd.concat([concat_stu_data, math_score], axis=1)
print (concat_stu_data2)
```
